```python
from sklearn.cluster import KMeans
import numpy as np
from matplotlib import pyplot as plt
```

```python
origin = np.loadtxt("heightweight.txt")
data = origin[:,1:]
female = origin[origin[:,0]==0, 1:]
male = origin[origin[:,0]==1, 1:]
plt.plot(female[:,0], female[:,1], 'bo', label='female')
plt.plot(male[:,0], male[:,1], 'r^', label='male')
```

```python
k = 2
pred = KMeans(n_clusters=k).fit(data)
centers = pred.cluster_centers_
labels = pred.labels_

lines = ["bo","r^"]
for i in range(k):
    class_i = data[labels==i]
    plt.plot(class_i[:,0], class_i[:,1], lines[i])
```
