```python
from sklearn.cluster import KMeans
import numpy as np
from matplotlib import pyplot as plt
```

```python
data = np.loadtxt("data1.txt")
k = 6
ax1 = plt.axes(projection='3d')
ax1.plot3D(data[:,0], data[:,1], data[:,2],'b^')
```

```python
lines = ["ro","g*","bx","ms","cp","kd"]
pred = KMeans(n_clusters=k).fit(data)
centers = pred.cluster_centers_
labels = pred.labels_
ax2 = plt.axes(projection='3d')
for i in range(k):
    class_i = data[labels==i]
    ax2.plot3D(class_i[:,0], class_i[:,1], class_i[:,2], lines[i])
```
